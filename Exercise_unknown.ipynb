## More about Random numbers

```python
import ROOT as rt
import numpy as np
```

```python
c1 = rt.TCanvas('c1', 'Random Fill', 200, 10, 700, 900)
c1.SetFillColor(18)

form1 = rt.TFormula('form1', 'abs(sin(x)/x)')
sqroot = rt.TF1('sqroot', 'x*gaus(0) + [3]*form1', 0, 10)
sqroot.SetParameters(10, 0, 1, 2)
rt.gPad.SetGridx()
rt.gPad.SetGridy()
sqroot.Draw()
c1.Draw()
```
